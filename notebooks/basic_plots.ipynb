```python
from mpl_toolkits.basemap import Basemap, shiftgrid, cm, addcyclic
import netCDF4 as nc
import numpy as np
import pylab as plt


plot_lat_range = [-30, 60]
plot_lon_range = [0, 90]
plt.figure(figsize=(6,6), dpi=200)
plt.clf()
plt.subplot(2,1,1)
newmap=Basemap(llcrnrlat=plot_lat_range[0],
               urcrnrlat=plot_lat_range[1],
               llcrnrlon=plot_lon_range[0],
               urcrnrlon=plot_lon_range[1],resolution='c')
newmap.drawparallels(np.arange(plot_lat_range[0], plot_lat_range[1], 30.), labels=[True,False,False,True])
newmap.drawmeridians(np.arange(plot_lon_range[0], plot_lon_range[1], 60.), labels=[True,False,False,True])
newmap.drawcoastlines(linewidth=0.5)
```

```python
newmap.pcolormesh(lons, lats, ozone[0,:,:], vmin=0, vmax=1e-8, cmap=plt.cm.Reds)
plt.colorbar()
plt.show()
```
